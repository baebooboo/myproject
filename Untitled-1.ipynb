```python
import re
def clean_column_name(col):
    match = re.match(r"^(\d{4})(?:\D*?)(\.\d+)?$", col)
    if match:
        return match.group(1) + (match.group(2) if match.group(2) else "")
    return col  # 매칭 안 되면 원래대로

marriage.columns = [clean_column_name(col) for col in marriage.columns]
marriage.head()
```

```python
# 숫자형 변환
marriage_f = marriage_f.apply(pd.to_numeric, errors='coerce')

# 시각화 코드 실행
plt.figure(figsize=(18, 8))
bar_width = 0.13
x = np.arange(len(marriage_f.index))

for i, year in enumerate(marriage_f.columns):
    offset = x + (i - len(marriage_f.columns)/2) * bar_width
    plt.bar(offset, marriage_f[year], width=bar_width, label=year)

plt.xticks(x, marriage_f.index.tolist(), rotation=45, ha='right')
plt.title("지역별 연도별 혼인 건수", fontsize=16)
plt.ylabel("혼인 건수")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title="연도", bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
